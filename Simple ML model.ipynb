```python
#Binarization of input.So that we can make sum of all 1s present and 
#filter it according to our threshold value b. 
X_binarized_train = X_train.apply(pd.cut,bins=2,labels=[1,0])
X_binarized_test = X_test.apply(pd.cut,bins=2,labels=[1,0])
```

```python
print("type of x_binarized_Train input: ",type(X_binarized_train))
print("type of x_binarized_Test input: ",type(X_binarized_test))
```

```python
#now to perform some operations we have to firstly convert it into numpy array.
X_binarized_train = X_binarized_train.values
X_binarized_test = X_binarized_test.values
```

```python
#checking if they are converted to numpy array
print("type of x_binarized_Train input: ",type(X_binarized_train))
print("type of x_binarized_Test input: ",type(X_binarized_test))
```

```python
#taking some random i i.e random row and checking if our model predicts
#it as malignant or beningn

from random import randint
i=randint(0,X_binarized_train.shape[0])

#b can take values from 1 to X_binarized_train.shape[1] i.e. here as 1 to 30
b=10
```

```python
#check what is ideal output and what our output predicts
if(np.sum(X_binarized_train[i,:])>=b):
    print("Model prediction is Malignant")
else:
    print("Model prediction is Beningn")
    
if(Y_train[i]==1):
    print("Actual output is Malignant")
else:
    print("Actual output is Beningn")
    

```

```python
#taking all the values of b i.e from 0 to 30 and taking the b 
#which is having maximum accuracy among them
maximum=0
for b in range(X_binarized_train.shape[1]+1):
    accurate_rows=0
    for x,y in zip(X_binarized_train,Y_train):
        y_pred = np.sum(x)>=b
        accurate_rows+=(y==y_pred)
    accuracy = accurate_rows/X_binarized_train.shape[0]
    print("for b = ",b,"\naccurate_Rows = ",accurate_rows,
          "\ntotal rows = ",X_binarized_train.shape[0],
          "\nAccuracy (accurate_rows/total_rows) = ",accuracy)
    maximum = max(maximum,accuracy)
    if maximum == accuracy:
        max_b = b
    
    print("====================================================================================")
    
print("Maximum acuracy for training dataset is:",maximum,
      "with b value as",max_b)
```

```python
#testing on test dataset
accurate_rows_test = 0
for x,y in zip(X_binarized_test,Y_test):
        y_pred = np.sum(x)>=max_b
        accurate_rows_test+=(y==y_pred)
accuracy = accurate_rows_test/X_binarized_test.shape[0]
print("for b = ",max_b,"\nAccurate_Rows = ",accurate_rows_test,
      "\ntotal rows = ",X_binarized_test.shape[0],
      "\ncuracy for testing dataset (accurate_rows/total_rows) = "
      ,accuracy)
    
```

```python
#Creating Confusion matrix:
print("total rows in testing dataset:",X_binarized_test.shape[0])
print("accurate rows in testing dataset: ",accurate_rows_test)
print("Inaccurate rows in testing dataset: ",
      X_binarized_test.shape[0] - accurate_rows_test)
```

```python
#creating confusion matrix where tp revers to the count of true positives,tn-true negatives
#fp - false positives and fn- false negatives
tp=0
tn=0
fp=0
fn=0
for i in range(X_binarized_test.shape[0]+1):    
    if((np.sum(X_binarized_test[i,:])>=max_b) and (Y_test[i]==1)):
        print("This is true postive")
        tp+=1
    
    if((np.sum(X_binarized_test[i,:])<=max_b) and (Y_test[i]==0)):
        print("This is true negative")
        tn+=1
    if((np.sum(X_binarized_test[i,:])>=max_b) and (Y_test[i]==0)):
        print("This is false positive")
        fp+=1
        
    if((np.sum(X_binarized_test[i,:])<=max_b) and (Y_test[i]==1)):
        print("This is false negative")
        fn+=1

```
